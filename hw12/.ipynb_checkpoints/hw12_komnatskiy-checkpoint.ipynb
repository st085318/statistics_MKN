```python
#2
data = pd.read_csv('Fish.csv')
data = data.drop(labels = [40],axis = 0)
y = data['Weight']

volume = data['Length1'] * data['Height'] * data['Width']
X = sm.add_constant(volume)
model = sm.OLS(y, X).fit()
plt.scatter(model.fittedvalues, model.resid)
plt.xlabel('fitted values')
plt.ylabel('resid')
plt.show()

y = model.resid ** 2
X = sm.add_constant(pd.concat((volume, volume * volume), axis=1))


model = sm.OLS(y, X).fit()
plt.scatter(volume, y, label='data')
plt.xlabel('volume')
plt.ylabel('e^2')
x = np.linspace(0, 5000, 1000)

def func(model_):
    a, b, c = model.params
    def fu(vol):
        return a + b * vol + c * vol ** 2
    return fu

plt.plot(x, func(model)(x), color = 'orange', label='prediction')
plt.legend()
plt.show()
model.fvalue
```

```python
#4
data = pd.read_csv('Advertising.csv')
y = data['sales']
X_s = sm.add_constant(pd.concat([data['radio'], 
               data['TV'] * data['radio'], 
               np.log(data['TV'])], axis = 1))
#print(X)
model_s = sm.OLS(y, X_s).fit()
rss_s = np.sum(model_s.resid ** 2)
y = data['sales']
X_l = sm.add_constant(pd.concat([data['TV'], data['radio'], 
              data['newspaper'], data['TV'] * data['radio'], 
               np.log(data['TV'])], axis = 1))
model = sm.OLS(y, X_l).fit()
rss_l = np.sum(model.resid ** 2)
delta_sample = rss_s - rss_l

deltas = []
boots_size = 10000
np.random.seed(42)
for i in range(boots_size):
    e_b = np.random.choice(model_s.resid, size=n)
    y_b = model_s.fittedvalues.values + e_b
    model = sm.OLS(y_b, X_s).fit()
    rss_s_b = np.sum(model.resid ** 2)
    model = sm.OLS(y_b, X_l).fit()
    rss_l_b = np.sum(model.resid ** 2)
    deltas.append(rss_s_b - rss_l_b)
    
(deltas > delta_sample).mean()
```

```python
#5
data = pd.read_csv('Fish.csv')
data = data.drop(labels = [40],axis = 0)
n = len(data)
k = 3
y = np.log(data['Weight'])
X = sm.add_constant(np.log(data[['Length1', 'Height', 'Width']]))
model = sm.OLS(y, X).fit()
_x = X.values
h = _x @ np.linalg.inv(_x.T @ _x) @ _x.T
diag = np.array([h[i][i] for i in range(n)])
leverage = np.where(diag > 3 * (k + 1)/n)[0]
for i in range(len(leverage)):
    leverage[i] = leverage[i] if leverage[i] < 40 else leverage[i] + 1
t_e = model.resid / (model.scale * (1 - diag)) ** 0.5
e = np.where(abs(t_e) > 3)[0]
for i in range(len(e)):
    e[i] = e[i] if e[i] < 40 else e[i] + 1
    
print(f'{leverage}\n{e}')
```
