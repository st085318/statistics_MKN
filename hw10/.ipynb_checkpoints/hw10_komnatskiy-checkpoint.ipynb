```python
import numpy as np
import pandas as pd
import scipy.stats as ss
import statsmodels.api as sm
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
```

```python
#2
data = pd.read_csv('view_proxy_randomization.csv')
def to_vasya(category):
    return category / 10 * 8 - 4

y = data['VIEW']
X = sm.add_constant(to_vasya(data['category']))
model = sm.OLS(y, X).fit()
print(model.summary(slim=True))
print(f"params {model.params}")
print(f"var: {model.scale ** 0.5}")
print(f"\n{model.params['const']} {model.params['category']} {model.scale ** 0.5}")

def func(a, b, w):
    return a + b * w

alpha = 0.05
q_category = 11
for category in range(q_category):
    c = data[data['category'] == category]['VIEW']
    n = len(c)
    t = ss.t(n - 1).ppf(1 - alpha / q_category / 2)
    up, down = c.mean() + c.std(ddof=1) / np.sqrt(n) * np.array([-t, t])
    plt.plot((category / 10 * 8 - 4, category / 10 * 8 - 4), (down, up), color='blue')

a = model.params['const']
b = model.params['category']
x = np.linspace(-4,4,1000)
plt.plot(x, func(a, b, x), color='orange', label='prediction')
plt.xlabel("w_vasya")
plt.ylabel("VIEW")
plt.legend()
plt.show()
```

**const:** среднее количество просмотров у людей из 5 категории(когда $\omega_{vasya}$=0)

**$\omega_{vasya}$:** (в коэфф. записан как category) если взять двух людей, у которых $\omega_{vasya}$ отличается на 1, то у человека с большей категорией, в среднем количество просмотров меньше на 1.207037

```python
#3
data = pd.read_csv('body_temp.csv')
y = data['body temperature']

ceneter_heart_rate = data['heart rate'] - data['heart rate'].mean()
X = sm.add_constant(pd.concat([data['sex'], ceneter_heart_rate], axis=1))
model = sm.OLS(y, X).fit()
print(model.summary(slim=True))
print(f"params {model.params}")
print(f"var: {model.scale ** 0.5}")


def func(avg_hr, a, b, c, x, sex=1):
    return a + b * sex + c * (x - avg_hr)


man = data[data['sex'] == 1]['heart rate']
woman = data[data['sex'] == 2]['heart rate']
plt.scatter(x=man, y=data[data['sex'] == 1]['body temperature'], c='orange', label='man')
plt.scatter(x=woman, y=data[data['sex'] == 2]['body temperature'], c='blue', label='woman')
x = np.linspace(50, max(data['heart rate']), 1000)
avg_hr = data['heart rate'].mean()
a = model.params['const']
b = model.params['sex']
c = model.params['heart rate']
plt.plot(x, func(avg_hr, a, b, c, x, 1), color='orange', label='prediction for men')
plt.plot(x, func(avg_hr, a, b, c, x, 2), color='blue', label='prediction for women')

plt.xlabel('heart rate')
plt.ylabel('body temperature')
plt.legend()
plt.show()
```

**const:** так как нельзя сделать sex = 0, то интерпритируем так: средняя температура мужчины со средней частотой сердцебиения равна 36.7316, а средняя температура женщины со средней частотой сердцебиения равна 36.8839
    
**sex:** То есть если взять мужчину и женщину с одинаковой частотой сердцебиения, то температура женщины в среднем будет больше на 0.152277 градуса

**heart rate:** То есть если взять двух людей одного пола, с частотой сердцебиения, отличающейся на 1, то температура человека с большей частотой в среднем будет на 0.013765 градуса выше 

```python
#4
data = pd.read_csv('birthweight.csv')
f_to_kg  = 0.453592
inch_to_cm = 2.54

n = len(data)


y = data['Birthweight'] * f_to_kg
c_gestation = data['Gestation'] - data['Gestation'].mean()
c_mheight = (data['mheight'] - data['mheight'].mean()) * inch_to_cm
smoker = data['smoker']
X = sm.add_constant(pd.concat([smoker, c_gestation, c_mheight], axis=1))
model = sm.OLS(y, X).fit()
print(model.summary(slim=True))
print(f"params {model.params}")
print(f"var: {model.scale ** 0.5}")
print(f"\n  {model.params['const'], model.params['smoker'], model.params['Gestation'], model.params['mheight'], model.scale ** 0.5}")
```

признаки - курение, время беременности, рост

**const:**
средний вес ребенка, которого вынашивала в течение средней продолжительности беременности, некурящая мать со средним ростом, будет составлять 3.456722 кг

**smoker:**
То есть если взять двух мам, у которых одна из которых курит, а другая нет, а остальные признаки совпадают, то веса ребенка некурящей матери в среднем будет на 0.3087 кг больше

**Gestation:**
То есть если взять двух мам, у которых время беремеенности отличалось на 1 неделю, а остальные признаки совпадают, то вес ребенка, мать которого вынашивала дольше в среднем будет на 0.1433 кг больше

**mheight:**
То есть если взять двух мам, у которых рост отличается на 1 см, а остальные признаки совпадают, то вес ребенка высокой мамы в среднем будет на 0.021737 кг больше

```python
#5
data = pd.read_csv('cars.csv')
y = np.log(data['Suggested Retail Price'])
X = sm.add_constant(np.log(data['Horsepower']))
model = sm.OLS(y, X).fit()
print(model.summary(slim=True))
print(f"params {model.params}")
print(f"var: {model.scale ** 0.5}")

print(f"\n{model.params['Horsepower']}")
def func(a, b, x):
    return a * (x ** b)

plt.scatter(x=data['Horsepower'], y=data['Suggested Retail Price'], c='blue', label='real cars')
x = np.linspace(0, max(data['Horsepower']), 1000)
a = np.exp(model.params['const'])
b = model.params['Horsepower']
plt.plot(x, func(a, b, x), color='orange', label='prediction')
plt.xlabel('Horsepower')
plt.ylabel('Suggested Retail Price')
plt.legend()
plt.show()
```

**Horsepower:**
То есть если взять две машины, у которых количество лошадинных сил отличается в n раз, то их цены в среднем будут отличаться в примерно $36.6 \times n^{0.15}$ ($36.579276681283034 \times n^{0.15227708427284437
}$) раз
